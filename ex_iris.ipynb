```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
import seaborn as sb
iris_df = sb.load_dataset('iris')
sb.pairplot(iris_df, hue='species', size=2.5) # hue là cột để phân loại 
```

```python
# số chiều của dữ liệu
iris_df.shape
```

```python
# thống kê dữ liệu cơ bản
iris_df.describe()
```

```python
# Hiển thị các quan sát từ dòng thứ 2 đến dòng thứ 5, mỗi quan sát liệt kê các giá trị dữ liệu từ cột thứ 1 đến cột thứ 4
iris_df.iloc[1:5, 0:4]
```

```python
# hiển thị tên các thuộc tính
print("Tên các thuộc tính là: ",iris_df.columns)
```

```python
# Tính giá trị trung bình, độ lệch chuẩn, giá trị lớn nhất và giá trị bé nhất của thuộc tính Petal.Length
iris_df["petal_length"].describe()
```

```python
soluong = iris_df["species"].value_counts()
from matplotlib import pyplot as plt
plt.bar(soluong.index, soluong.values) # index là tên các loài hoa, values là số lượng
plt.show()
```

```python
# Sử dụng pairs plot để nhìn tổng quan dữ liệu
import seaborn as sns
sns.pairplot(iris_df, hue="species")
plt.show()
```

```python
# Sử dụng heatmap để kiểm tra xem các thuộc tính nào có thể có mối liên hệ với nhau
# plt.figure(figsize=(10,11))
sns.heatmap(iris_df.corr(), annot=True)
plt.plot()
```

```python
# Sử dụng đồ thị histogram để so sánh phân phối của 4 thuộc tính trong dữ liệu
iris_df.hist(edgecolor='blue', linewidth=1.2) # edgecolor là màu viền, linewidth là độ dày viền
fig=plt.gcf() # gcf là get current figure
fig.set_size_inches(12,6)
plt.show()
```

```python
# Sử dụng đồ thị boxplot để so sánh phân phối 4 thuộc tính của các loại hoa
plt.figure(figsize=(15,10)) # tạo ra 1 figure có kích thước 15x10
plt.subplot(2,2,1) # tạo ra 1 subplot 2x2, vị trí 1
sns.boxplot(x='species',y='sepal_length',data=iris_df) # vẽ boxplot
plt.subplot(2,2,2) # tạo ra 1 subplot 2x2, vị trí 2
sns.boxplot(x='species',y='sepal_width',data=iris_df) # vẽ boxplot
plt.subplot(2,2,3) # tạo ra 1 subplot 2x2, vị trí 3
sns.boxplot(x='species',y='petal_length',data=iris_df) # vẽ boxplot
plt.subplot(2,2,4) # tạo ra 1 subplot 2x2, vị trí 4
sns.boxplot(x='species',y='petal_width',data=iris_df) # vẽ boxplot
```
