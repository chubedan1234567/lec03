```python
soluong = iris_df["species"].value_counts()
from matplotlib import pyplot as plt
plt.bar(soluong.index, soluong.values) # index l tn cc loi hoa, values l s lng
plt.show()
```

```python
# S dng pairs plot  nhn tng quan d liu
import seaborn as sns
sns.pairplot(iris_df, hue="species")
plt.show()
```

```python
# S dng heatmap  kim tra xem cc thuc tnh no c th c mi lin h vi nhau
# plt.figure(figsize=(10,11))
sns.heatmap(iris_df.corr(), annot=True)
plt.plot()
```

```python
# S dng  th histogram  so snh phn phi ca 4 thuc tnh trong d liu
iris_df.hist(edgecolor='blue', linewidth=1.2) # edgecolor l mu vin, linewidth l  dy vin
fig=plt.gcf() # gcf l get current figure
fig.set_size_inches(12,6)
plt.show()
```

```python
# S dng  th boxplot  so snh phn phi 4 thuc tnh ca cc loi hoa
plt.figure(figsize=(15,10)) # to ra 1 figure c kch thc 15x10
plt.subplot(2,2,1) # to ra 1 subplot 2x2, v tr 1
sns.boxplot(x='species',y='sepal_length',data=iris_df) # v boxplot
plt.subplot(2,2,2) # to ra 1 subplot 2x2, v tr 2
sns.boxplot(x='species',y='sepal_width',data=iris_df) # v boxplot
plt.subplot(2,2,3) # to ra 1 subplot 2x2, v tr 3
sns.boxplot(x='species',y='petal_length',data=iris_df) # v boxplot
plt.subplot(2,2,4) # to ra 1 subplot 2x2, v tr 4
sns.boxplot(x='species',y='petal_width',data=iris_df) # v boxplot
```
